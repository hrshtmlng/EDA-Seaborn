```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df1 = df[["total_bill","tip","size"]]
```

```python
## Correlation
df1.corr()
```

```python
#Heatmap

sns.heatmap(df1.corr())
```

```python
## JOINPLOT

# Univariate Analysis

sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
#Dist plot
sns.distplot(df['tip'])
```

 Categorical plots
    #boxplot

```python
#count plot

sns.countplot(x='sex',data=df)
```

```python
#bar plot
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
#Box plot
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
#Violin Plot
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
